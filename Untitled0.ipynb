```python
import yfinance as yf
import matplotlib.pyplot as plt

# 設定台積電的股票代碼 (2330.TW 是台灣股票代碼)
ts_code = '2330.TW'

# 取得台積電的歷史股價資料
df = yf.download(ts_code, period='1y') # 下載過去一年的資料

# 顯示前幾行資料確認
print(df.head())
```

```python
# 繪製台積電的收盤價走勢圖

import matplotlib.font_manager as fm
from matplotlib import rc

# 檢查系統中是否有支援中文的字體，如果沒有，則安裝一個
# 這裡以'Taipei Sans TC Beta'為例，你也可以使用其他你喜歡的字體
# 如果你的系統中沒有，可能需要手動安裝，或者更換為其他可用的中文字體
font_name = fm.FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc').get_name() # 或者你電腦裡有的中文字體路徑

rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False # 解決負號亂碼問題

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='收盤價')
plt.title('台積電 (2330.TW) 過去一年股價走勢圖')
plt.xlabel('日期')
plt.ylabel('股價 (新台幣)')
plt.grid(True)
plt.legend()
plt.show()
```
